```python
#Storing URL and attempt retrieval of page using request function
URL = 'https://sg.finance.yahoo.com/quote/%5EGSPC/history?period1=1575158400&period2=1596240000&interval=1wk&filter=history&frequency=1wk&includeAdjustedClose=true'

response1 = requests.get(URL)
stat_code = response1.status_code

content1 = response1.content
print(stat_code) #status code 200 = success!
```

```python
#Assigning parser to BS parsing through content
parser = BeautifulSoup(content1, 'html.parser')
body = parser.body

#Locating the main data body through .find operator
histdata = body.find("div", id = "Col1-1-HistoricalDataTable-Proxy")

#Creating empty list to add contents through scraping techniques
histheader = []
histdate = []
histclose = []

#Adding all information located in page's table of interest.
histheader.append("S&P500 " + histdata.find("span", {'data-reactid':'37'}).text)
histheader.append("S&P500 " + histdata.find("th", {'data-reactid':'44'}).text[0:5])

#Looping the data through the table in script contents to add into lists.
for hdate in histdata.find_all("td", class_="Py(10px) Ta(start) Pend(10px)"):  #Looking for all the data in "Date" column
    histdate.append(hdate.text)

for hclose in histdata.find_all("td", class_="Py(10px) Pstart(10px)"): # Looking for all the data in "Close" column
    histclose.append(hclose.text)    
    
#Sample checking if data is correctly added into list
print (histheader)
print (histdate[:10])
print (histclose[:20])
#Seems good!
```

## 6) The Auto Regressive Integrated Moving Average Model (ARIMA) model for time-series analysis

* <strong>What is the ARIMA model work?</strong>

<p>The ARIMA model is a simple regressive model that utilizes time-series data & statistical concepts to train past data to predict future outcomes. It is commonly used to forecast future stock-prices or company performances based on past results.</p>

* <strong>How does the ARIMA model work?</strong>

<p>ARIMA models predict future values based on past values. It uses past lagged moving averages to smooth out time-series data.A weight is placed on past data which are based on how recent the data is.It can also be summarized through below points:</p> 


* Autoregression (AR): referring to a model that is auto-changing due to the variables in itself.
* Integrated (I): differencing its raw-observations such that data-points are the distance between value and its older value on a different time-point.
* Moving average (MA):  utilizing error-terms which is used to calculate the future-terms. It smooths out and removes any random movements from a time-series.
