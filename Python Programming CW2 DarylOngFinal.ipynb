```python
#Storing URL and attempt retrieval of page using request function
URL = 'https://sg.finance.yahoo.com/quote/%5EGSPC/history?period1=1575158400&period2=1596240000&interval=1wk&filter=history&frequency=1wk&includeAdjustedClose=true'
headers ={'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36'}

response1 = requests.get(URL, headers=headers)
stat_code = response1.status_code

content1 = response1.content
print(stat_code) #status code 200 = success!
```

```python
#Assigning parser to BS parsing through content
parser = BeautifulSoup(content1, 'html.parser')
body = parser.body

#Locating the main data body through .find operator
histdata = body.find("div", id = "Col1-1-HistoricalDataTable-Proxy")

#Creating empty list to add contents through scraping techniques
histheader = []
histdate = []
histclose = []

#Adding all information located in page's table of interest.
histheader.append("S&P500 " + histdata.find("span", {'data-reactid':'37'}).text)
histheader.append("S&P500 " + histdata.find("th", {'data-reactid':'44'}).text[0:5])

#Looping the data through the table in script contents to add into lists.
for hdate in histdata.find_all("td", class_="Py(10px) Ta(start) Pend(10px)"):  #Looking for all the data in "Date" column
    histdate.append(hdate.text)

for hclose in histdata.find_all("td", class_="Py(10px) Pstart(10px)"): # Looking for all the data in "Close" column
    histclose.append(hclose.text)    
    
#Sample checking if data is correctly added into list
print (histheader)
print (histdate[:10])
print (histclose[:20])
#Seems good!
```

### <a id='5.1'> 5.1) Test Driven Development

<p>For the upcoming simple function to normalize the cells, I've added this TDD segment additionally from my CW1 to test this normalize function. In summmary, a TDD test is as follows:</p>

* Checks functionality of code prior to writing.

* More likely to catch bugs when tests are executed.

* Provides a clearer picture on the objectives in lines of codes.

```python
#Simple Function Code

def normalize(x,y):
    return x/y
```

```python
#TDD test codes

import unittest

class TDD_Test(unittest.TestCase):

    def test_normalize_exist(self):
        self.assertIsNotNone(normalize)
        
    def test_normalize_1(self):
        self.assertEqual(normalize(1 , 2), 0.5)
        
    def test_normalize_2(self):
        self.assertEqual(normalize(1 , 5), 0.2)
        
    def test_normalize_3(self):
        self.assertEqual(normalize(10 , 100), 0.1)
    
unittest.main(argv =['ignore' , '-v'], exit= False)
```

```python
import re #importing our REGEX functions

#using Regular-expression syntax, we define the critical keywords which most-likely affects the market trends below.
keywords = r'\b(?:first death|emergency|highest single-day death toll|outbreak)\b' 

df_keyword_bear_event = pd.DataFrame(columns=['Date','Event-Global']) #defining first filtered dataframe
df_keyword_bear_event2 = pd.DataFrame(columns=['Date','Event-Global']) #defining re-scrubbed filtered dataframe

for i in df_event['Event-Global']:
    if re.findall(keywords, i):  #Using REGEX to find keywords
        keysearch = df_event[df_event['Event-Global'] == i].index
        df_i_event = df_event.loc[keysearch] #.loc to find the index in original event dataframe
        df_keyword_bear_event = df_keyword_bear_event.append(df_i_event,ignore_index=True)

#Below lines re-scrubs the keywords to form events only native to U.S news.
keywords_lv2 = r'\b(?:US|WHO)\b'
        
for i in df_keyword_bear_event['Event-Global']:
    if re.findall(keywords_lv2, i): #Using REGEX to find keywords
        keysearch2 = df_keyword_bear_event[df_keyword_bear_event['Event-Global'] == i].index
        df_i_event2 = df_keyword_bear_event.loc[keysearch2] #.loc to find the index in original event dataframe
        df_keyword_bear_event2 = df_keyword_bear_event2.append(df_i_event2,ignore_index=True)
        
df_keyword_bear_event2
```

#### Dickey Fuller Test

<p>In statistics, a fairly common test to evaluate the fitness of a dataset is the <strong>augmented Dickey-Fuller test</strong>(or known as the ADF test). This test tests the 'Unit root' which indicates if the data is 'stationary'. More about stationarity will be explained below.<br> To summarize, in this test we will get the p-value of this particular dataset. If the p-value is above critical value which is 0.05, we cannot reject the null hypothesis.</p>

* <strong>Null-Hypothesis</strong> - if critical value below 0.05 we <strong>will</strong> reject null-hypothesis meaning that the series IS stationary.

* <strong>Alternate-Hypothesis</strong> - if critical value above 0.05 we <strong>cannot</strong> reject null-hypothesis meaning that the series IS NOT stationary.

```python
#Implementing the dickey-fuller test.

from statsmodels.tsa.stattools import adfuller

def adf_test(data):
    df_adtest = adfuller(data, autolag = 'AIC')
    print('1. ADF : ', df_adtest[0])
    print('2. P-Value : ', df_adtest[1])
    print('3. No. of Lags : ', df_adtest[2])
    print('4. No of observations used for ADF Regression & Critical Values calculations : ', df_adtest[3])
    print('5. Critical Values : ')
    for key,val in df_adtest[4].items():
        print('\t', key,': ', val)

print(adf_test(df_nasdaq_daily['Nasdaq-Close']))
```

<p>From above test, we can observe that the P-value is greater than that of 0.05 which surpassed that critical value. This means that we cannot reject the null hypothesis and proves that the data-set we have is non-stationary. </p>

### Background on Stationarity forecasting — ARIMA models

<p>How is stationarity important with respect to time-series forecasting? Our topic at hand requires ample understanding to this concept as we move forward. Stationary time series is defined as <b>"When the mean & variance are constant over time"</b>.
In a nutshell, when calculating time-series models, the more the data's movement revolves around the mean & variance of the overall dataset, it is easier to make predictions with said data. This also implies that the data is no 'trend' existing.</p>

#### Solving the issue of a Non-stationary data - Differencing.

<p>Shifting our focus back to the data-set at hand, we will implement a method of transforming a non-stationary time series into a stationary one. This method is called "Differencing", which takes the difference between the current time period & the previous time period. If this value fails to revolve around the constant mean, we will repeat the method again until we get a stationary series.</p>

<p>For the following lines I will attempt to transform the dataset using differencing and implement the AD-fuller test again to test the P-value.</p>

```python
#Implementing differencing transform for one shift period.

df_nasdaq_daily['Nasdaq-Close_diff'] = df_nasdaq_daily['Nasdaq-Close'] - df_nasdaq_daily['Nasdaq-Close'].shift(1)
df_nasdaq_daily['Nasdaq-Close_diff'] = df_nasdaq_daily['Nasdaq-Close_diff'].dropna()
df_nasdaq_daily['Nasdaq-Close_diff'].plot()
```

* p - This parameter is associated to the auto-regressive aspect of the model which notes the weight of past lag values to make predictions.

* d - Associated with the amount of differencing applied to a time-snap of value points through the data-set.

* q - q is the parameter that signifies the moving average part of the model. The higher this parameter is, the more our algorithm associates the moving average of the training data.

* s - seasonal indicator for the length of seasonal change in data.

### Possible Caveats of analysis.

<p>Albeit that above comparisons yielded some insights on how our predictions could react in actual world situations in terms of a time-series analysis, it has a few blindsides. Below listed are a few observed:</p>

* Normalized Stationary data - The transformed data has actually skewed the actual value but kept the trend lines only. In terms of these time-point specific analysis, it has caused the accuracy of the price movements to drop.

* Extreme outliers - Due to the nature of ARIMA models, it generally predicts better of values that are more in range of the mean. Hence this model tend to over or underestimate extreme values(or outliers).
